```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/drive/MyDrive/Neural Network /week-2/Week-2 Assignment, EDA/gym_members_exercise_tracking.csv'
data = pd.read_csv(file_path)

# Display basic information about the dataset
print("Data Info:")
data.info()
print("\nFirst 5 Rows of Data:")
print(data.head())
print("\nDataset Shape:", data.shape)
print("\nSummary Statistics:")
print(data.describe())

# Set visualization style
sns.set(style="whitegrid")

# Plot distributions of key numerical variables
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle("Distributions of Key Numerical Variables", fontsize=16)

# Plot each distribution in the grid
sns.histplot(data['Age'], kde=True, ax=axes[0, 0], color="skyblue").set(title="Age Distribution")
sns.histplot(data['Weight (kg)'], kde=True, ax=axes[0, 1], color="salmon").set(title="Weight Distribution")
sns.histplot(data['Height (m)'], kde=True, ax=axes[0, 2], color="green").set(title="Height Distribution")

sns.histplot(data['Max_BPM'], kde=True, ax=axes[1, 0], color="purple").set(title="Max BPM Distribution")
sns.histplot(data['Calories_Burned'], kde=True, ax=axes[1, 1], color="orange").set(title="Calories Burned Distribution")
sns.histplot(data['Session_Duration (hours)'], kde=True, ax=axes[1, 2], color="teal").set(title="Session Duration Distribution")

sns.histplot(data['Water_Intake (liters)'], kde=True, ax=axes[2, 0], color="blue").set(title="Water Intake Distribution")
sns.histplot(data['Workout_Frequency (days/week)'], kde=True, ax=axes[2, 1], color="brown").set(title="Workout Frequency Distribution")
sns.histplot(data['BMI'], kde=True, ax=axes[2, 2], color="gray").set(title="BMI Distribution")

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit title
plt.show()

# Calculate and visualize the correlation matrix
# Select only numeric columns for correlation calculation
numerical_data = data.select_dtypes(include=['number'])  # Select only numeric columns
correlation_matrix = numerical_data.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Gym Members' Exercise Data", fontsize=16)
plt.show()
```
